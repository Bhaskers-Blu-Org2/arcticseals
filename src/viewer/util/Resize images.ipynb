## Resize images 25%

```python
import os
import pathlib
from PIL import Image
```

```python
# create small-color
colorSmallDir = 'E:\\noaa\\small-color'
new_width = 1644
new_height = 1096

#define the path
currentDir = pathlib.Path('F:\\')
for currentFile in currentDir.glob("*.jpg"):
    img = Image.open(currentFile)
    # 6576px x 4384px -> 1644px x 1096
    img = img.resize((new_width, new_height), Image.ANTIALIAS)
    
    # currentFile = CHESS_FL12_C_160421_210008.597_COLOR-8-BIT.JPG
    # outputFile = E:\noaa\small-color\CHESS_FL12_C_160421_210008.597_COLOR-8-BIT.JPG
    currentFilename = os.path.basename(currentFile)
    outputFile = os.path.join(colorSmallDir, currentFilename)
    img.save(outputFile)
```
